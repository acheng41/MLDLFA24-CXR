```python
epi_15_metrics = {}

# Iterate through the keys and append the data
for key in epi15_metrics_8.keys():
    # Combine the arrays along the first axis (or appropriate axis for your data)
    epi_15_metrics[key] = np.concatenate(
        (epi15_metrics_8[key], epi15_metrics_15[key]), axis=0
    )

# Save the combined data into a new .npz file
np.savez("episodic_eta_1_5_metrics.npz", **epi_15_metrics)
```
